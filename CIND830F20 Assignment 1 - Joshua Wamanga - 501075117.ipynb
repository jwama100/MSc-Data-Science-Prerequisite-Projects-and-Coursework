```python
if side1 > 0 and side2 > 0 and side3 > 0:
    if side1 == side2 == side3:
        print("You have an Equilateral Triangle since all side are congruent.")
    else:
        print("Your triangle is not Equilateral since the sides aren't congruent.")
else:
    print("Error: The input is invalid, please input positive integers into the program.")
```
