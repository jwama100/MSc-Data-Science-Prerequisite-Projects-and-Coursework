**a)** Complete the methods of the following `Stack` class according to their description

```python
class Stack:
  def __init__(self):
    """ Initialize a new stack """
    self.elements = []
    self._top = -1
    self._size = 0

  def push(self, new_item):
    """ Append the new item to the stack """
    self.elements.append(new_item)
    self._top += 1
    self._size += 1
    self.elements[self._top] = new_item

  def pop(self):
    """ Remove and return the last item from the stack """
    oldItem = self.elements[self._top]
    self._top -= 1
    self._size -= 1
    return self.elements.pop()

  def size(self):
    """ Return the total number of elements in the stack """
    return len(self.elements)

  def is_empty(self):
    """ Return True if the stack is empty and False if it is not empty """
    if self.elements == 0:
        return True
    else:
        return False

  def peek(self):
    """ Return the element at the top of the stack or return None if the stack is empty """
    if self.elements != 0:
        return self.elements[self._top]
    else:
        return None

testStack = Stack()  # Testing stack object
testStack.push(2) # Testing operations
testStack.push(4)
testStack.push(6)
print(testStack.size())
print(testStack.pop()) # Testing operations
print(testStack.peek())
print(testStack.size())
```
